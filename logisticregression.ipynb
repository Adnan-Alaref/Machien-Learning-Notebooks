# **◍ Step 1: Import Library.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

# **◍ Step 2: Build MOdel.**

```python
def logistic_model(case):
  """
    Train and evaluate a logistic regression model on two datasets:
    - Linearly separable dataset
    - Non-linearly separable dataset

    Parameters:
    - case (bool): If True, use the linearly separable dataset. Otherwise, use the non-linearly separable dataset.
  """
  use_linearly_separable = case
  # Data
  if use_linearly_separable:    # linearly separable
      print("Using linearly separable dataset...\n")
      X = np.array([[0, 2], [4, 1], [3, 10], [9, 1], [8, 7]])
      y = np.array([0, 0, 1, 1, 1])
  else:       # NOT linearly separable
      print("Using non-linearly separable dataset...\n")
      X = np.array([[0, 2], [3, 10], [9, 1], [8, 7]])
      y = np.array([0, 1, 1, 0])

  # Build model
  model = LogisticRegression()
  model.fit(X,y)

  # Make prediction
  # SciKit uses default threshold = 0.5
  y_pred = model.predict(X) # Predicted labels
  y_proba = model.predict_proba(X) # Predicted probabilities for each class

  # Display Model Insights
  print("Model Parameters:")
  print(f"Intercept: {model.intercept_[0]:.4f}") # return list with one item so acsess index[0]
  print(f"Coefficients:\n{model.coef_}\n")


  # Evaluation
  print("Model Predictions:")
  print(f"Ground truth labels: {y}")
  print(f"Predicted labels: {y_pred}")
  print(f"Predicted probabilities for all classes:\n{y_proba}")
  print(f"Correctly classified: {np.count_nonzero(y_pred==y)} out of {y.size}")
  print(f"Probabilities for class 0: {y_proba[:,0]}")
  print(f"Probabilities for class 1: {y_proba[:,1]}\n\n")


  # Drowing
  # Get parameters of mx+c equation / hyperplance
  b = model.intercept_[0]
  w1 ,w2 = model.coef_.T
  m ,c = -w1/w2 , -b/w2          # Don't work if w2=0

  plt.plot(X ,m * X + c ,color = 'b' ,linewidth = 3)  # decision boundary

  indexces_class_0 = np.argwhere(y==0).reshape(-1) # argwhere return 2D array (n,1) then reshape -> convert it to 1D array
  X_index_class_0 = X[indexces_class_0]            # Way1 extract X values corresponding to class 0
  plt.plot(X_index_class_0[:,0] ,X_index_class_0[:,1] ,"o" ,color = 'b' ,label = "Class0")


  indexces_class_1 = np.argwhere(y==1).reshape(-1)                   # argwhere return 2D array (n,1) then reshape -> convert it to 1D array
  x_index_class_1 = np.array([X[idx] for idx in indexces_class_1])  # Way2 extract X values corresponding to class 1 return -> list convert it to array
  plt.plot(x_index_class_1[:,0] ,x_index_class_1[:,1] ,"s",color = "r" ,label = "Class1")

  plt.xlabel("x")
  plt.ylabel("y")
  plt.grid()
  plt.legend(prop = {"size":10} ,loc = "best")
  plt.title("Logistic Regression Decision Boundary" ,fontweight= "bold" ,fontsize = 12)
  plt.show()
```

```python
# Call the function to test
logistic_model(True)
```

```python
# Call the function to test
logistic_model(False)
```
